```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 6, 18]
y = [3, 10, 12, 20]

plt.plot(x, y,color="r",marker="o",linestyle="dotted")

plt.xticks(np.arange(min(x), max(x)+1, 1))
plt.yticks(np.arange(min(y), max(y)+1, 1))

plt.xlabel("x")
plt.ylabel("y")

plt.title("graph")

plt.show()
```

```python
f = open("new.csv")
ans = f.readlines()
print(ans)

x=[]
for i in ans:
    lis = i.split(',')
    lis = list(map(float,lis))
    x.append(lis)

plt.plot(x[0],x[1],color="r",marker="o")
plt.title("none")

plt.show()
```

```python
x = [1,2,3,4,5]
y1=[5,10,5,30,2]
y2=[10,2,31,4,50]

plt.plot(x,y1,label="line1",marker='o',color="r")
plt.plot(x,y2,label="line2",marker="s",color="g")
plt.legend()
plt.show()
```
